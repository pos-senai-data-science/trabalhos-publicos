```python
from google.colab import files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Passo 1: Carregamento do csv
uploaded = files.upload()
```

```python
# Passo 2: Ler os dados usando o pandas
file_name = next(iter(uploaded))
data = pd.read_csv(file_name, sep=',')
```

```python
# Label
data['Label'] = data.iloc[:, 3].apply(lambda x: 1 if x == 'Masculino' else 0)
data
```

```python
data.iloc[:, 1:3] = data.iloc[:, 1:3].replace('175', '1,75', regex= True)
data.iloc[:, 1:3] = data.iloc[:, 1:3].replace(',', '.', regex= True)
data.iloc[:, 1:3] = data.iloc[:, 1:3].replace('kg', '', regex= True)

data
```

```python
data.iloc[:, 1:3] = data.iloc[:, 1:3].astype(float)
data
```

```python
data.dtypes
```

```python
# Criar um scatter plot usando os dados do DataFrame
plt.scatter(data.iloc[:, 1], data.iloc[:, 2], c= data.iloc[:, 4], cmap='coolwarm')

# Adicionar rótulos e título
plt.xlabel('Altura')
plt.ylabel('Peso')
plt.title('Gráfico de Dispersão')


# Mostrar o gráfico
plt.show()
```

```python
# Bibliotecas do SVM
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
def normaliza(x, max, min):
  return (x - min)/(max - min)
```

```python
# Separação dos dados em X e y
X = data.iloc[:, 1:3].values
y = data.iloc[:, 4].values

max_ = X.max()
min_ = X.min()

X = normaliza(X, max_, min_)
```

```python
# Teste e treino
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=9)
```

```python
# Criação da instância
model = KNeighborsClassifier()
```

```python
# Criação da instância de treino
model.fit(X_train, y_train)
```

```python
# Predição dos dados de test
y_pred = model.predict(X_test)
```

```python
# Cálculo da acurácia
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
